```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("KPMG_1.xlsx",sheet_name='CustomerDemographic', skiprows=1)
df
```

```python
df.info()
```

```python
df['default'].value_counts()
```

```python
df = df.drop('default',axis=1)
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['last_name']=df['last_name'].fillna(method='ffill')
```

```python
df['DOB']=df['DOB'].fillna(method='ffill')
```

```python
df['job_title'].value_counts()
```

```python
df['job_title']=df['job_title'].fillna(method='bfill')
```

```python
df['job_industry_category'].value_counts()
```

```python
df['job_industry_category']=df['job_industry_category'].fillna(method='ffill')
```

```python
df['job_industry_category'].value_counts().plot(kind='bar')
plt.show()
```

```python
df['tenure'].mean()
```

```python
df['tenure']=df['tenure'].fillna(df['tenure'].mean())
```

```python
df['gender'].value_counts()
```

```python
df['gender'] = df['gender'].replace('Femal','Female').replace('M','Male').replace('F', 'Female').replace('U','Unspecified')
```

```python
df['gender'].value_counts()
```

```python
df['gender'].value_counts().plot(kind='bar')
plt.show()
```

```python
submission = pd.DataFrame({
        "customer_id": df["customer_id"],
        "first_name": df["first_name"],
        "last_name": df["last_name"],
        "gender": df["gender"],
        "past_3_years_bike_related_purchases": df["past_3_years_bike_related_purchases"],
        "DOB": df["DOB"],
        "job_title": df["job_title"],
        "job_industry_category": df["job_industry_category"],
        "wealth_segment": df["wealth_segment"],
        "deceased_indicator": df["deceased_indicator"],
        "owns_car": df["owns_car"],
        "tenure": df["tenure"]
    })
submission.to_csv('./submission_CustomerDemographic.csv', index=False)
print(submission.head(10))
```
